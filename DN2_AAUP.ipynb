```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
aaup = pd.read_csv('aaup_data.csv')
```

***
### Q1. Distribution of Colleges by Type
***

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.set_style("whitegrid")
plt.figure(figsize=(8,6))
order=['I', 'IIA', 'IIB']
ax = sns.countplot(x="Type", data=aaup, palette="Set2", order=order)
plt.title("Number of Colleges by Type", fontsize=16)
plt.xlabel("Type", fontsize=14)
plt.ylabel("Count", fontsize=14)
ax.patches[0].set_facecolor('#8da0cb')
ax.patches[2].set_facecolor('#66c2a5')
plt.show()
```

***
### Q2. Salary Distribution across College Types
***

```python
import seaborn as sns
sns.set_style("ticks")
sns.kdeplot(data=aaup, x="Avg. Salary (all ranks)", hue="Type", fill=True, common_norm=False, alpha=.5)
```

***
### Q3. Number of Professors across College Types
***

```python
import seaborn as sns
sns.set_style("ticks")
sns.kdeplot(data=aaup, x="Number of Faculty (all ranks)", hue="Type", fill=True, common_norm=False, alpha=.5)
```

***
### Q4. Average Salary vs Average Compensation
***

```python
sns.set_style('whitegrid')
sns.jointplot(data=aaup, x='Avg. Compensation (all ranks)', y='Avg. Salary (all ranks)', hue='Type')
```

***
### Q5. Total Expenditure on Professors in each type of Institute
***

```python
aaup['Avg. Salary (full prof.)'] = pd.to_numeric(aaup['Avg. Salary (full prof.)'], errors='coerce')
aaup['Avg. Compensation (full prof.)'] = pd.to_numeric(aaup['Avg. Compensation (full prof.)'], errors='coerce')
aaup['Number of Full Prof.'] = pd.to_numeric(aaup['Number of Full Prof.'], errors='coerce')

aaup['Avg. Salary (asso. prof.)'] = pd.to_numeric(aaup['Avg. Salary (asso. prof.)'], errors='coerce')
aaup['Avg. Compensation (asso. prof.)'] = pd.to_numeric(aaup['Avg. Compensation (asso. prof.)'], errors='coerce')
aaup['Number of Asso. Prof.'] = pd.to_numeric(aaup['Number of Asso. Prof.'], errors='coerce')

aaup['Avg. Salary (asst. prof.)'] = pd.to_numeric(aaup['Avg. Salary (asst. prof.)'], errors='coerce')
aaup['Avg. Compensation (asst. prof.)'] = pd.to_numeric(aaup['Avg. Compensation (asst. prof.)'], errors='coerce')
aaup['Number of Asst. Prof.'] = pd.to_numeric(aaup['Number of Asst. Prof.'], errors='coerce')

aaup['Total Expenditure'] = (aaup['Avg. Salary (full prof.)']+aaup['Avg. Compensation (full prof.)'])*aaup['Number of Full Prof.'] + (aaup['Avg. Salary (asso. prof.)']+aaup['Avg. Compensation (asso. prof.)'])*aaup['Number of Asso. Prof.'] + (aaup['Avg. Salary (asst. prof.)']+aaup['Avg. Compensation (asst. prof.)'])*aaup['Number of Asst. Prof.']
aaup['Total Expenditure'] = pd.to_numeric(aaup['Total Expenditure'], errors='coerce')

plt.figure(figsize=(10,10))
ax = sns.boxplot(x='Type', y='Total Expenditure', data=aaup, palette='Set2', order=['I', 'IIA', 'IIB'])
ax.patches[0].set_facecolor('#fc8d62')
ax.patches[1].set_facecolor('#66c2a5')
ax.patches[2].set_facecolor('#8da0cb')
ax = plt.gca()
ax.set_xlabel('Institution Type')
ax.set_ylabel('Total Expenditure (in Millions)')
```
